```python
import os
import subprocess
import cv2

VIDEO_NAMES = ["walk_front","walk_front_2","walk_side", "walk_45","walk_330"]#["LWF"]#, "LWS"]
MODEL_NAME = "LHM-1B-HF"  # Ensure this matches actual model dir
# IMG_PATH = "./inputs/images/1.jpg"
# IMG_PATH = "./inputs/images/lxh.jpg"
IMG_PATH = "./inputs/images/lxh45.jpg"
# IMG_PATH = "./inputs/images/T_frontview.jpg"
IMG_NAME = os.path.basename(IMG_PATH).split(".")[0]
if not os.path.exists(IMG_PATH):
    raise FileNotFoundError(f"Image not found at path: {IMG_PATH}")
def resize_by_shorter_side(img, target_short_side=512):
    if img is None:
        raise ValueError(f"Image not found or could not be read atpath: {IMG_PATH}")
    h, w = img.shape[:2]
    if h < w:
        new_h = target_short_side
        new_w = int(w * (target_short_side / h))
    else:
        new_w = target_short_side
        new_h = int(h * (target_short_side / w))
    return cv2.resize(img, (new_w, new_h), interpolation=cv2.INTER_LINEAR)

new_img = resize_by_shorter_side(cv2.imread(IMG_PATH))
cv2.imwrite(IMG_PATH, new_img)
```

```python
import json



def fix_trans_location(input_dir, output_dir):
    
    # === Ensure output folder exists ===
    os.makedirs(output_dir, exist_ok=True)

    # === Load all JSON file paths, sorted ===
    json_files = sorted(f for f in os.listdir(input_dir) if f.endswith(".json"))

    # === Extract reference translation from first file ===
    with open(os.path.join(input_dir, json_files[0]), 'r') as f:
        first_data = json.load(f)
    trans_ref = first_data['trans']

    # === Process and save each JSON with fixed translation ===
    for fname in json_files:
        input_path = os.path.join(input_dir, fname)
        output_path = os.path.join(output_dir, fname)

        with open(input_path, 'r') as f:
            data = json.load(f)

        data['trans'] = trans_ref  # Replace translation
        with open(output_path, 'w') as f:
            json.dump(data, f, indent=2)

    print(f"Processed {len(json_files)} files. Output saved to '{output_dir}'.")
```

```python
fixed_tran_folder = []
for folder in out_list:
    fixed_out_path = os.path.join(folder,"fixed_trans")
    print(f"[INFO] Creating fixed translation directory: {fixed_out_path}")
    os.makedirs(fixed_out_path, exist_ok=True)
    fix_trans_location(folder, fixed_out_path)
    fixed_tran_folder.append(fixed_out_path)
```

```python
for curr in fixed_tran_folder:
    subprocess.run([
        "bash", "inference.sh", MODEL_NAME, IMG_PATH, curr
    ], check=True)
```
